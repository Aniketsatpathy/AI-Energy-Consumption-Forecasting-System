```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("data/raw/household_power_consumption.txt", sep=';')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)

# Convert target column to numeric
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')

# Drop missing values
df = df.dropna()

df.head()
```

```python
df['timestamp'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Sort by time
df = df.sort_values('timestamp')

df[['timestamp', 'Global_active_power']].head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['Global_active_power'][:1000])
plt.title("Raw Energy Consumption (Sample)")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.grid()
plt.show()
```

```python
# Set index
df = df.set_index('timestamp')

# Convert to hourly data
df_hourly = df.resample('h').mean()

df_hourly = df_hourly.reset_index()

df_hourly.head()
```

## 🔧 Feature Engineering

To improve forecasting performance, we create additional features:

### ⏳ Lag Features
- lag_1 → previous hour
- lag_24 → same hour previous day
- lag_168 → same hour previous week

👉 Helps model remember past patterns

---

### 📊 Rolling Features
- rolling_mean_24 → average of last 24 hours
- rolling_std_24 → variability

👉 Helps model understand trends

---

### 🧠 Time Features
- hour
- day_of_week
- month

👉 Captures seasonal behavior

```python
df_feat = df_hourly.copy()

df_feat['energy'] = df_feat['Global_active_power']

# Time features
df_feat['hour'] = df_feat['timestamp'].dt.hour
df_feat['day_of_week'] = df_feat['timestamp'].dt.dayofweek
df_feat['month'] = df_feat['timestamp'].dt.month

# Lag features
df_feat['lag_1'] = df_feat['energy'].shift(1)
df_feat['lag_24'] = df_feat['energy'].shift(24)
df_feat['lag_168'] = df_feat['energy'].shift(168)

# Rolling features
df_feat['rolling_mean_24'] = df_feat['energy'].rolling(24).mean()
df_feat['rolling_std_24'] = df_feat['energy'].rolling(24).std()

# Drop missing values
df_feat = df_feat.dropna()

df_feat.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df_feat['energy'][:500])
plt.title("Processed Energy Data")
plt.grid()
plt.show()
```

## 📊 Key Insights

- Energy consumption follows strong daily patterns
- Peak usage occurs during evening hours
- Weekly trends are clearly visible
- Short-term history (lag_1) plays a major role in prediction
- Rolling averages help smooth noisy data

---

## 🧠 Conclusion

Time-series forecasting benefits significantly from:
- Temporal features
- Historical lag values
- Trend-based statistics

These features are later used in machine learning models to predict future energy consumption.
